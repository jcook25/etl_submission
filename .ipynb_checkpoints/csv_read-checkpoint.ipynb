```python
import json
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
file = "2005_data.csv"
data_05_df = pd.read_csv(file)
```

```python
data_05_slice = data_05_df[["resident_status", "sex", "place_of_death_and_decedents_status", "manner_of_death", "method_of_disposition"]]
data_05_slice
```

```python
df = data_05_slice.dropna()
df = df.rename(columns={"resident_status":"resident_status",
                        "sex":"sex",
                        "place_of_death_and_decedents_status":"place_of_death",
                        "manner_of_death":"manner_of_death",
                        "method_of_disposition":"method_of_disposition"})
df
```

```python
dfc = df.reset_index()
dfc = dfc.drop(columns='index')
dfc.dtypes
```

```python
dfcc = dfc.astype({"manner_of_death": int})
dfcc
```

```python
dfcc.dtypes
```

```python
dfsb = dfcc["manner_of_death"]

dfsb.value_counts()
```

```python
# Connect to local database
rds_connection_string = f"postgres:{password}@localhost:5432/etl_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check for tables
engine.table_names()
```

```python
# Use pandas to load csv converted DataFrame into database
dfcc.to_sql(name='data_05', con=engine, if_exists='fail', index=False)
```
